```python
import pybullet as p
import pybullet_data
import numpy as np
import matplotlib.pyplot as plt

# connect without graphics
p.connect(p.DIRECT)

# load default models
p.setAdditionalSearchPath(pybullet_data.getDataPath())

# gravity
p.setGravity(0, 0, -9.81)

# load ground plane
plane = p.loadURDF("plane.urdf")

# load cube 1 meter above ground
cube = p.loadURDF("cube.urdf", [0,0,1])

z_positions = []
time_steps = []

dt = 1/240

for i in range(1000):

    pos, orn = p.getBasePositionAndOrientation(cube)
    z_positions.append(pos[2])
    time_steps.append(i*dt)

    p.stepSimulation()

p.disconnect()

# plot height vs time
plt.plot(time_steps, z_positions)
plt.xlabel("time (s)")
plt.ylabel("cube height (m)")
plt.title("PyBullet gravity test")
plt.show()
```
